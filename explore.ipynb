```python
import json
import os
import pandas as pd 
```

Config

```python
restaurant_name = "souvla"
source_name = "yelp"
folder = f"./data_raw/{restaurant_name}/{source_name}/"
```

```python
filenames = os.listdir(folder)
filename = "dataset_yelp-scraper_2023-05-31_02-19-37-173.json"
```

Load

```python
with open(f"{folder}/{filename}", "rb") as f:
    data = json.load(f)
```

View

```python
for restaurant in data:
    print("Restaurant: ", restaurant["name"], restaurant["directUrl"])
    print("Reviews: ", len(restaurant["reviews"]))
    print(" ")
```

```python
restaurant = data[0]

clean_reviews_df = pd.DataFrame(columns=["rating", "review"])
for review in restaurant["reviews"]:
    clean_reviews_df = clean_reviews_df.append({"rating": review["rating"], "review": review["text"].replace("\n", "")}, ignore_index=True)
clean_reviews_df.sort_values(by=["rating"], inplace=True, ascending=True)

clean_reviews = ""
for index, row in clean_reviews_df.iterrows():
    clean_reviews += f"Rating: {row['rating']}\tReview: {row['review']}\n"

clean_path = f"./data_clean/{restaurant_name}/{source_name}/"
if not os.path.exists(clean_path):
    os.makedirs(clean_path)

with open(f"{clean_path}/clean_reviews.txt", "w") as f:
    f.write(clean_reviews)
```
